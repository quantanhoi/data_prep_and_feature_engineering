```python
import pandas as pd
df = pd.read_csv('./housing.csv')
print(df.head(2))
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
# 2) Z-score normalise *total_rooms*
scaler = StandardScaler()
df['total_rooms_z'] = scaler.fit_transform(df[['total_rooms']])  # (x-µ)/σ



```

```python
# 3) plot — two histograms side-by-side, with mean lines
fig, ax = plt.subplots(1, 2, figsize=(10, 4))
# before
ax[0].hist(df['total_rooms'], bins=100, color='blue', edgecolor='black')
ax[0].set_title('total_rooms  (raw)')
ax[0].set_xlabel('rooms')
ax[0].set_ylabel('count')
mean_raw = df['total_rooms'].mean()
ax[0].axvline(mean_raw, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_raw:.0f}')
ax[0].legend()

# after
ax[1].hist(df['total_rooms_z'], bins=100, color='orange', edgecolor='black')
ax[1].set_title('total_rooms  (z-score)')
ax[1].set_xlabel('z value')
ax[1].set_ylabel('count')
mean_z = df['total_rooms_z'].mean()
ax[1].axvline(mean_z, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_z:.2f}')
ax[1].legend()

plt.tight_layout()
plt.savefig('z_score_housing.png', dpi=300)
plt.show()
```
